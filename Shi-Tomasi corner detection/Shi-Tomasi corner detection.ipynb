```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image and convert to grayscale
img = cv2.imread('burj.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Shi-Tomasi corner detection: Only strong, sharp corners
corners = cv2.goodFeaturesToTrack(
    gray,
    maxCorners=40,           # Lower to show only important ones
    qualityLevel=0.1,        # Higher = only stronger corners
    minDistance=20           # Prevent clustering
)
corners = np.intp(corners)


for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 3, (0, 255, 0), -1, lineType=cv2.LINE_AA)

# Display result
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Strong Shi-Tomasi Corners Only")
plt.axis('off')
plt.show()
```
